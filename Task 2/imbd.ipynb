### Compute mean rating and count of films for each category, for my opinion i this this is the way to gain  more informations from tha dataset 


```python
df['Genre Mean'] = df.groupby('Genre')['Rating'].transform(lambda x: x.mean())
df['Genre Count'] = df.groupby('Genre')['Rating'].transform('count')

df['Director Mean'] = df.groupby('Director')['Rating'].transform(lambda x: x.mean())
df['Director Count'] = df.groupby('Director')['Rating'].transform('count')

df['Actor 1 Mean'] = df.groupby('Actor 1')['Rating'].transform(lambda x: x.mean())
df['Actor 1 Count'] = df.groupby('Actor 1')['Rating'].transform('count')

df['Actor 2 Mean'] = df.groupby('Actor 2')['Rating'].transform(lambda x: x.mean())
df['Actor 2 Count'] = df.groupby('Actor 2')['Rating'].transform('count')

df['Actor 3 Mean'] = df.groupby('Actor 3')['Rating'].transform(lambda x: x.mean())
df['Actor 3 Count'] = df.groupby('Actor 3')['Rating'].transform('count')

df['Name Mean'] = df.groupby('Name')['Rating'].transform(lambda x: x.mean())
df['Name Count'] = df.groupby('Name')['Rating'].transform('count')
```

```python
df['Name'] = df['Name Mean'] + df['Name Count']
df['Actor 1'] = df['Actor 1 Mean'] + df['Actor 1 Count']
df['Actor 2'] = df['Actor 2 Mean'] + df['Actor 2 Count']
df['Actor 3'] = df['Actor 3 Mean'] + df['Actor 3 Count']
df['Director'] = df['Director Mean'] + df['Director Count']
df['Genre'] = df['Genre Mean'] + df['Genre Count']
```

```python
df.drop(columns = ['Genre Mean', 'Genre Count', 'Director Mean', 'Director Count', 'Actor 1 Mean', 'Actor 1 Count', 
                   'Actor 2 Mean', 'Actor 2 Count', 'Actor 3 Mean', 'Actor 3 Count', 'Name Mean', 'Name Count'], inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
```

```python
X = df.drop(columns=['Rating'])
y = df['Rating']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2) Score: {r2:.2f}")
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2) Score: {r2:.2f}")
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2) Score: {r2:.2f}")
```
