```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["Year"].unique(), df['Duration'].unique(), df['Votes'].unique()
```

```python

df['Year'] = df['Year'].str.replace('(', '').str.replace(')', '').astype(float)
```

```python
df['Duration'] = df['Duration'].str.replace(' min', '').astype(float)
```

```python
df["Votes"]=df["Votes"].str.replace(",","")
df["Votes"]=df["Votes"].str.replace("$5.16M","516",regex=False)
```

```python
df.Votes = df.Votes.astype(float)
```

```python
df.Duration =  df["Duration"].fillna(df["Duration"].median())
```

```python
df.Rating =  df["Rating"].fillna(df["Rating"].median())
```

```python
df.Votes =  df["Votes"].fillna(df["Votes"].median())
```

```python
df.isnull().sum()
```

### imputing Genre, year, director and actors columns it doesn't make sense and if we do that, it will make fake prediction

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.Genre.value_counts()
```

### for Genre column we will separate it so instead of a movie that has a genre of drama, crime in one row this will be in two rows movie with genre drama and the same movie with genre crime

```python
df["Genre"]=df["Genre"].str.split(",").explode("Genre").reset_index(drop=True)
```

```python
df
```

```python
df.Genre.value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
genre_counts = df['Genre'].value_counts()

# Plot the genre counts
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('From th top genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
director_ratings = df.groupby('Director')['Rating'].mean().sort_values()

# Plot the bar chart of directors and their average ratings
plt.figure(figsize=(14, 8))
director_ratings.plot(kind='barh', color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Director')
plt.title('Average Ratings of Films by Director')
plt.grid(axis='x')
plt.show()
```
